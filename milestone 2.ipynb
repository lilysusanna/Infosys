```python
def merge_diarization_stt(diarization, transcripts):
    merged = []
    for speaker, start, end in diarization:
        # find overlapping STT segments
        text = " ".join([t['text'] for t in transcripts if t['start'] >= start and t['end'] <= end])
        merged.append(f"[{speaker}]: {text}")
    return merged

diarized_transcript = merge_diarization_stt(segments, transcripts)
for line in diarized_transcript:
    print(line)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

def summarize_text(text, max_length=150):
    inputs = tokenizer("summarize: " + text, return_tensors="pt", truncation=True)
    summary_ids = model.generate(inputs.input_ids, max_length=max_length, min_length=40, length_penalty=2.0, num_beams=4)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
full_transcript = " ".join(diarized_transcript)
summary = summarize_text(full_transcript)
print("====Summary====")
print(summary)
```
